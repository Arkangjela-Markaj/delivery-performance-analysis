```python
from sklearn.datasets import load_iris
import pandas as pd

# We'll create our own realistic delivery dataset instead
import numpy as np

np.random.seed(42)
n = 10000

df = pd.DataFrame({
    'trip_id': range(1, n+1),
    'driver_id': np.random.randint(1, 101, n),
    'zone': np.random.choice(['North', 'South', 'East', 'West', 'Downtown'], n),
    'hour': np.random.randint(0, 24, n),
    'trip_duration_min': np.random.normal(25, 10, n).clip(5, 90).round(1),
    'distance_miles': np.random.normal(5, 2, n).clip(0.5, 20).round(2),
    'delivery_success': np.random.choice([1, 0], n, p=[0.85, 0.15]),
    'delay_min': np.random.exponential(5, n).round(1),
    'vehicle_type': np.random.choice(['Van', 'Bike', 'Car'], n, p=[0.4, 0.3, 0.3])
})

print("Dataset created! Shape:", df.shape)
print(df.head())
```

```python
# Explore the data
print("=== Dataset Info ===")
print(df.info())

print("\n=== Missing Values ===")
print(df.isnull().sum())

print("\n=== Basic Statistics ===")
print(df.describe().round(2))
```

```python
# Explore the data
print("=== Dataset Info ===")
print(df.info())

print("\n=== Missing Values ===")
print(df.isnull().sum())

print("\n=== Basic Statistics ===")
print(df.describe().round(2))
```

```python
# Feature engineering
df['speed_mph'] = (df['distance_miles'] / (df['trip_duration_min'] / 60)).round(2)
df['time_of_day'] = pd.cut(df['hour'], 
                            bins=[0, 6, 12, 17, 21, 24], 
                            labels=['Night', 'Morning', 'Afternoon', 'Evening', 'Late Night'],
                            right=False)
df['is_delayed'] = (df['delay_min'] > 10).astype(int)

print("New columns added!")
print(df[['trip_id', 'speed_mph', 'time_of_day', 'is_delayed']].head(10))
```

```python
# Business Question 1: Which zones have the highest failure rates?
zone_analysis = df.groupby('zone').agg(
    total_trips=('trip_id', 'count'),
    successful=('delivery_success', 'sum'),
    avg_delay=('delay_min', 'mean'),
    avg_duration=('trip_duration_min', 'mean')
).round(2)

zone_analysis['failure_rate_%'] = ((1 - zone_analysis['successful'] / zone_analysis['total_trips']) * 100).round(2)
zone_analysis = zone_analysis.sort_values('failure_rate_%', ascending=False)

print("=== Delivery Performance by Zone ===")
print(zone_analysis)
```

```python
# Visualize Zone Performance
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Chart 1: Failure rate by zone
colors = ['#d73027' if x == zone_analysis['failure_rate_%'].max() 
          else '#4575b4' for x in zone_analysis['failure_rate_%']]
axes[0].bar(zone_analysis.index, zone_analysis['failure_rate_%'], color=colors)
axes[0].set_title('Delivery Failure Rate by Zone (%)')
axes[0].set_xlabel('Zone')
axes[0].set_ylabel('Failure Rate (%)')
axes[0].set_ylim(0, 25)

# Chart 2: Average delay by zone
axes[1].bar(zone_analysis.index, zone_analysis['avg_delay'], color='#74add1')
axes[1].set_title('Average Delay by Zone (minutes)')
axes[1].set_xlabel('Zone')
axes[1].set_ylabel('Avg Delay (min)')

plt.tight_layout()
plt.savefig('zone_performance.png', dpi=150)
plt.show()
print("Chart saved!")
```

```python
# Business Question 2: Performance by time of day
time_analysis = df.groupby('time_of_day', observed=True).agg(
    total_trips=('trip_id', 'count'),
    failure_rate=('delivery_success', lambda x: (1 - x.mean()) * 100),
    avg_delay=('delay_min', 'mean'),
    avg_speed=('speed_mph', 'mean')
).round(2)

time_analysis = time_analysis.sort_values('failure_rate', ascending=False)

print("=== Performance by Time of Day ===")
print(time_analysis)
```

```python
# Business Question 2: Top and bottom performing drivers
driver_analysis = df.groupby('driver_id').agg(
    total_trips=('trip_id', 'count'),
    success_rate=('delivery_success', 'mean'),
    avg_delay=('delay_min', 'mean'),
    avg_speed=('speed_mph', 'mean')
).round(2)

driver_analysis['success_rate'] = (driver_analysis['success_rate'] * 100).round(2)

# Top 5 drivers
print("=== TOP 5 DRIVERS ===")
print(driver_analysis.sort_values('success_rate', ascending=False).head())

# Bottom 5 drivers
print("\n=== BOTTOM 5 DRIVERS ===")
print(driver_analysis.sort_values('success_rate', ascending=True).head())
```

```python
# Visualize driver performance
fig, ax = plt.subplots(figsize=(12, 5))

# Plot all drivers as scatter
ax.scatter(driver_analysis['avg_delay'], 
           driver_analysis['success_rate'],
           alpha=0.6, color='#4575b4', s=60)

# Highlight top 3 and bottom 3
top3 = driver_analysis.sort_values('success_rate', ascending=False).head(3)
bot3 = driver_analysis.sort_values('success_rate').head(3)

ax.scatter(top3['avg_delay'], top3['success_rate'], 
           color='green', s=100, zorder=5, label='Top drivers')
ax.scatter(bot3['avg_delay'], bot3['success_rate'], 
           color='red', s=100, zorder=5, label='Bottom drivers')

ax.set_title('Driver Performance: Success Rate vs Average Delay')
ax.set_xlabel('Average Delay (min)')
ax.set_ylabel('Success Rate (%)')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('driver_performance.png', dpi=150)
plt.show()
print("Chart saved!")
```

```python
# Business Question 3: Vehicle type performance
vehicle_analysis = df.groupby(['vehicle_type', 'time_of_day']).agg(
    total_trips=('trip_id', 'count'),
    success_rate=('delivery_success', 'mean'),
    avg_delay=('delay_min', 'mean')
).round(2)

vehicle_analysis['success_rate'] = (vehicle_analysis['success_rate'] * 100).round(2)

print("=== Performance by Vehicle Type and Time of Day ===")
print(vehicle_analysis)
```

```python
# Final visualization - heatmap of success rate by vehicle and time
pivot = df.groupby(['vehicle_type', 'time_of_day'], observed=False)['delivery_success'].mean() * 100
pivot = pivot.unstack().round(1)

fig, ax = plt.subplots(figsize=(10, 5))
sns.heatmap(pivot, annot=True, fmt='.1f', cmap='RdYlGn', 
            ax=ax, vmin=80, vmax=95,
            linewidths=0.5)
ax.set_title('Delivery Success Rate (%) by Vehicle Type and Time of Day')
ax.set_xlabel('Time of Day')
ax.set_ylabel('Vehicle Type')

plt.tight_layout()
plt.savefig('vehicle_heatmap.png', dpi=150)
plt.show()
print("Chart saved!")
```
